# **Impressões na tela de um usuário**

> Existem diversos métodos de imprimir algo na tela de um usuário em C++. No começo, estaremos utilizando a [biblioteca iostream](https://cplusplus.com/reference/iostream/), que utiliza o comando **cout** para fazer a impressão. O comando 'cout' vem da linguagem C, que é a base do C++. 'Cout' é uma abreviação de 'Character Output'. Para utilizar esse comando em um código, iremos importar a biblioteca e criar a função principal.

```cpp
#include <iostream> // Para fazer a importação da biblioteca

int main(){ // Comando que irá gerar a função principal de seu código
std::cout << "Mensagem" << std::endl; // Imprime "Mensagem" e faz uma quebra de linha
return 0;  // Retorna 0 para indicar que o programa terminou com sucesso
}
```
> Vamos desmembrar a função e mostrar seu funcionamento passo a passo

```cpp
#include <iostream> // Este é o pré-processador do C++. Ele inclui a biblioteca iostream no seu programa.

int main() { // Esta é a função principal do seu programa. Todo programa em C++ começa a ser executado pela função main.

    // "int" é o tipo de retorno da função. Neste caso, a função main retorna um valor inteiro.
    // "main()" é o nome da função.
    // "{}" é o corpo da função, onde o código é escrito.

    std::cout << "Mensagem"; // Este é o comando de saída da biblioteca iostream. Ele imprime a string "Mensagem" na tela.

    std::endl; // Este é o comando para fazer uma quebra de linha. No entanto, ele precisa ser usado em conjunto com std::cout.

    return 0; // Este é o valor de retorno da função main. Quando o programa termina com sucesso, a função main retorna 0.
} // Fim da função principal


```

> # **Variáveis**
Uma variável em C++ é um nome dado a um local de armazenamento que nosso programa pode manipular. Cada variável em C++ tem um tipo específico, que determina o tamanho e o layout da memória da variável, a faixa de valores que ela pode armazenar e o conjunto de operações que podem ser aplicadas a ela.

> **Existem diversos tipos de variáveis em C++. Dentre eles, `float`, `double` e `int` são comumente utilizados para armazenar números. Veja abaixo alguns exemplos de como declarar e inicializar esses tipos de variáveis:**
- `int` : Este é um tipo de dado integral que armazena números inteiros.
```cpp
int Numero = 1; // Declaração de uma variável do tipo int
```
- `float` :  Este é um tipo de dado de ponto flutuante que pode armazenar números reais, ou seja, números que têm uma parte fracionária.
```cpp
float Numero2 = 2.1; // Declaração de uma variável do tipo float
```
- `double` : Este é outro tipo de dado de ponto flutuante que pode armazenar números reais. No entanto, ele tem o dobro da precisão de um float.
```cpp
double Numero3 = 3.2; // Declaração de uma variável do tipo double
```

> **Além dos tipos numéricos, C++ também suporta outros tipos de dados. Aqui estão alguns exemplos:**
- `String` : Este tipo é usado para armazenar uma sequência de caracteres ou texto. Por exemplo:
```cpp
std::string Texto = "Olá, mundo!"; // Declaração de uma variável do tipo string
```
- `bool` : Este é um tipo booleano que pode armazenar true (verdadeiro) ou false (falso). Por exemplo:
```cpp
bool VerdadeiroOuFalso = true; // Declaração de uma variável do tipo bool
```
- `char` : Este tipo é usado para armazenar um único caractere. Por exemplo:
```cpp
char Letra = 'A'; // Declaração de uma variável do tipo char
```

# **Estruturas de Controle**

> *Após entendermos sobre variáveis, podemos avançar para as estruturas de controle em C++. Elas são fundamentais para controlar o fluxo do programa. Aqui estão alguns exemplos:*

> - `if` : Esta é uma estrutura de controle condicional que executa um bloco de código se uma condição for verdadeira.
```cpp
int Numero = 10;
if (Numero > 5) {
    std::cout << "O número é maior que 5." << std::endl;
}
```
- `else` : Usado em conjunto com `if`, o `else` permite que executemos um bloco de código se a condição do `if` for falsa.
```cpp
int Numero = 4;
if (Numero > 5) {
    std::cout << "O número é maior que 5." << std::endl;
} else {
    std::cout << "O número é menor ou igual a 5." << std::endl;
}
```
- `for` : Este é um loop que executa um bloco de código um número específico de vezes.
```cpp
for (int i = 0; i < 5; i++) {
    std::cout << "Estamos no loop número " << i << std::endl;
}
```
`do`: Esta é uma estrutura de controle que executa um bloco de código pelo menos uma vez e depois repete a execução enquanto uma condição for verdadeira. Por exemplo:
```cpp
int i = 0;
do {
    std::cout << "Estamos no loop número " << i << std::endl;
    i++;
} while (i < 5);
 /*Neste exemplo, o bloco de código dentro do do será executado pelo menos uma vez, independentemente
 da condição no while. Depois disso, ele continuará executando enquanto a condição i < 5 for verdadeira.*/
```
> Essas são as estruturas de controle básicas em C++. Cada uma delas tem suas próprias características e usos, dependendo do que você precisa para o seu programa.
